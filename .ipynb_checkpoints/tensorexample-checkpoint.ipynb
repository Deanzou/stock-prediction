```python
#b=tf.expand_dims(b,-1)
tf.abs(tf.sign(a)).eval()
padding.eval()
b=tf.where(tf.equal(b, 0), padding, tf.cast(b,tf.float32))
tf.nn.softmax(b, 1).eval()
d=tf.expand_dims(tf.argmax(tf.nn.softmax(b, 1),1),-1).eval()
c.eval()
tf.equal(d,tf.cast(c,tf.int64))
```
